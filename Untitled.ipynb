```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.cluster import KMeans 
from sklearn.metrics import silhouette_score 
```

```python
iris = pd.read_csv("IRIS.csv")
```

```python
iris.head()
```

```python
iris.info()
```

```python
iris_ct = pd.crosstab(index = iris['species'],columns = 'count')
```

```python
iris_ct
```

```python
x = iris.iloc[:,[0,1,2,3]].values
```

```python
sns.boxenplot(x = 'species', y = 'petal_length', data = iris)
plt.show()
```

```python
sns.violinplot(x = 'species', y = 'petal_length', data = iris)
plt.show()
```

```python
sns.set_style('whitegrid')
sns.pairplot(iris, hue = 'species')
plt.show()
```

```python
distortions = []
K = range(1,10)

for k in K:
    kmeansmodel = KMeans(n_clusters = k)
    kmeansmodel.fit(x)
    distortions.append(kmeansmodel.inertia_)
```

```python
plt.figure(figsize = (16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method Showing the Optimal CLusters')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

```python
y_kmeans
```

```python
x[y_kmeans == 2, 0]
```

```python
#Visualising the clusters
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 20, c = 'purple', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 20, c = 'orange', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 20, c = 'green', label = 'Iris-virginica')

#Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 70, c = 'red', label = 'Centroids')

plt.legend()
plt.show()
```

```python
x = iris.iloc[:,[0,1,2,3]].values
```

```python
y = iris.iloc[:,4].values
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
# KMeans 
import random
random.seed(116)

kmeans = KMeans(n_clusters = 3)
y_pred = kmeans.fit_predict(x_test)

from sklearn.metrics import accuracy_score 
accuracy = accuracy_score(y_test,y_pred)

print('Accuracy for KMeans Clustering: ', accuracy)
```

```python
from sklearn.linear_model import LogisticRegression 
logmodel = LogisticRegression()
logmodel.fit(x_train, y_train)
```

```python
y_pred = logmodel.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix 
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score 
print('Accuracy for logistic regression: ', accuracy_score(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier_knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2) # p = 2 indicates euklidian distance
classifier_knn.fit(x_train,y_train)
```

```python
y_pred = classifier_knn.predict(x_test)
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy for KNeighborsClassifier: ', accuracy_score(y_test,y_pred))
```
